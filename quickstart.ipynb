```python
from img2vec_pytorch import Img2Vec
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Initialize Img2Vec with CPU
img2vec = Img2Vec(cuda=False)
```

```python
img_dir = 'example/test_images/'

img = Image.open(img_dir + 'cat.jpg')

img
```

```python
vec = img2vec.get_vec(img, tensor=True)
```

```python
vec.shape
```

```python
embedding = np.array(vec[0,:,0,0])
embedding = embedding/embedding.max()
```

```python
fig, ax = plt.subplots(1,1,figsize=(15,5))

index = [i for i in range(0, len(embedding))]
ax.bar(index, embedding);
```
